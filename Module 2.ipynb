## Task 1 - Plotting
* Plot the values of occupancy percentage versus number of weeks since release, for all movies in the dataset. 
* Plot the number of other releases in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.
* Plot the number of shows in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.

```python
import matplotlib.pyplot as plt
import pandas as pd
dataframe = pd.read_csv("Data/CinemaData.csv")
X1 = dataframe.iloc[:, 12]
Y1 = dataframe.iloc[:, 8]
plt.scatter(X1, Y1, color = "blue")
plt.xlabel("No_of_weeks_since_release")
plt.ylabel("Occupancy_Percentage")
plt.title("CinemaData")
plt.show()
Y2 = dataframe.iloc[:, 9]
X21 = dataframe.iloc[: ,8]
plt.scatter(X21, Y2, color = "red")
plt.xlabel("Occupancy_Percentage")
plt.ylabel("No_of_other_releases_in_the_week")
plt.title("CinemaData")
plt.show()
X22 = [dataframe.iloc[:, 5]*52]
plt.scatter(X22, Y2, color = "green")
plt.xlabel("Weeks_until_the_end_of_lifetime")
plt.ylabel("No_of_other_releases_in_the_week")
plt.title("CinemaData")
plt.show()
Y3 = dataframe.iloc[:, 11]
X31 = dataframe.iloc[:, 8]
plt.scatter(X31, Y3)
plt.xlabel("Occupancy_Percentage")
plt.ylabel("No_of_shows_in_the_week")
plt.title("CinemaData")
plt.show()
X32 = [dataframe.iloc[:, 5]*52]
plt.scatter(X32, Y3, color = "purple")
plt.title("CinemaData")
plt.xlabel("Weeks_until_the_end_of_lifetime")
plt.ylabel("No_of_shows_in_the_week")
plt.show()
```
